# **Project Summary**

## **Key findings**

- Học sinh nam ở độ tuổi 21–22 có điểm G3 giảm mạnh, trong khi nữ giới duy trì xu hướng ổn định hơn
- Sau 18 tuổi, mối quan hệ studytime → điểm số gần như biến mất
- Nam và Nữ không học hiệu quả theo cùng một cách: điểm số của Nam giới vượt trội Nữ giới ở mức học thấp (<5h), ở mức học cao (>10h) thì ngược lại
- Việc có hay không được gia đình hỗ trợ học tập (famsup) không làm thay đổi đáng kể điểm số
- Internet không có tác động đồng nhất – hiệu quả phụ thuộc vào goout và freetime

## **Limitations**

**Dataset Limitations:**
- Kích thước mẫu không đồng đều giữa các nhóm (ví dụ: rất ít học sinh 21-22 tuổi, Medu/Fedu=0), ảnh hưởng độ tin cậy thống kê ở một số phân tích.
- Dữ liệu cross-sectional (cắt ngang) nên chỉ thấy tương quan (correlation) chứ chưa khẳng định nhân quả (causation).
- Thiếu thông tin chi tiết về cách sử dụng Internet (học tập vs giải trí), chất lượng hỗ trợ gia đình (famsup), và nguyên nhân nghỉ học (absences).

**Analysis Limitations:**
- Chưa sử dụng phương pháp nâng cao như XGBoost, Neural Networks, hoặc stacking ensemble cho bài toán dự đoán rượu bia.
- Chưa thực hiện feature selection chi tiết (RFE, LASSO) hay hyperparameter tuning tối ưu (Grid Search, Bayesian Optimization) cho các mô hình machine learning.
- Một số kiểm định thống kê (như ANOVA cho Age Groups) không có ý nghĩa thống kê dù biểu đồ cho thấy xu hướng, có thể do sample size nhỏ.
- Dù đã phân tích tương tác 2 chiều, nhưng chưa xây dựng mô hình hồi quy đa biến đầy đủ (multiple regression) với các interaction terms để kiểm soát đồng thời nhiều yếu tố.

**Scope Limitations:**
- Chỉ phân tích 2 môn học (Toán và tiếng Bồ Đào Nha) của học sinh Bồ Đào Nha, chưa thể khẳng định kết quả áp dụng rộng rãi cho các môn khác hay quốc gia khác.
- Không xem xét tác động của giáo viên, chất lượng trường học, hoặc các yếu tố môi trường xã hội rộng hơn.
- Chưa phân tích tác động dài hạn của các yếu tố lên sự nghiệp học tập sau này.
- Chưa xem xét các yếu tố tâm lý, sức khỏe tinh thần có thể ảnh hưởng đến học tập và thói quen sinh hoạt do thiếu các feature tương ứng trong dữ liệu.

## **Future Direction**

**Cải thiện dữ liệu**
- Mở rộng phạm vi nghiên cứu: hiện tại bộ dữ liệu chỉ được khảo sát trên 2 môn Toán và Bồ Đào Nha của 2 trường trung học $\rightarrow$ Cần mở rộng thu nhập dữ liệu từ nhiều quốc gia để kiểm chứng xem liệu các yếu tố có tác động giống nhau hay không.
- Nghiên cứu theo chiều dọc: thay vì chỉ phân tích dữ liệu tại một thời điểm, theo dõi và phân tích hồ sơ học sinh xuyên suốt các năm của học sinh trung học để xem rõ mối quan hệ nhân quả.

**Advanced Modeling**
- Áp dụng mô hình học máy nâng cao: thử các thuật toán Ensemble như XGBoost, CatBoost. Các thuật toán này có khả năng xử lý mất cân bằng dữ liệu và nắm được các mối quan hệ phi tuyến phức tạp tốt hơn Decision Tree hay Logistic Regression.

**Practical Deployment**
- Hệ thống Cảnh báo sớm: Phát triển Dashboard thời gian thực dành cho nhà trường. Hệ thống sẽ dùng điểm số `G1` và các feature trên để tự động nhận diện và gắn cờ những học sinh có nguy cơ trượt môn cao $\rightarrow$ Giúp giáo viên can thiệp kịp thời.
- Cá nhân hoá lộ trình hỗ trợ: Sử dụng Clustering để nhóm các học sinh có cùng đặc điểm rủi ro và thiết kế các chương trình tư vấn tâm ly hoặc phụ đạo được cá nhân hoá riêng cho từng nhóm để có thể ủng hộ về mặt tinh thần cho học sinh.
