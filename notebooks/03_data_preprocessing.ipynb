```python
import pandas as pd

df = pd.read_csv("../data/raw/student-combine.csv")
```

## **Feature Engineering**

### Tạo cột `toal_alc` từ `Dalc` và `Walc`

- Do độ tương quan của `Dalc` và `Walc` tương đối mạnh nên ta sẽ gộp thành cột `total_alc` cho các mục đích phân tích
- Ta sẽ tính trung bình 2 biến giá trị nhưng do `Dalc` thì sẽ là từ thứ 2 - thứ 6 còn `Walc` là thứ 7, Chủ nhật nên ta sẽ có công thức:
$$
total\_alc = \frac{Dalc * 5 + Walc * 2}{7}
$$

```python
df['total_alc'] = round((df['Dalc'] * 5 + df['Walc'] * 2 ) / 7, 2)  # Tạo cột total_alc theo công thức trên và làm tròn 2 chữ số
df['total_alc']
```

### Tạo cột `parent_edu` từ `Medu` và `Fedu`

- Dựa vào phần trước, ta thấy mối tương qua của `Medu` và `Fedu` khá cao cho nên ta sẽ tạo ra biến giá trị trung bình để tổng hợp yếu tố nền tảng gia đình phục vụ cho các nhiệm vụ phân tích liên quan

```python
df['parent_edu'] = (df['Medu'] + df['Fedu']) / 2    # Lấy giá trị trung bình
```

## **Processing Outliers**

Dựa vào phần **Data Exploration**, ta thấy có outlier ở cột `absences` (giá trị rất lớn như 75) $\rightarrow$ ta xử lý bằng cách winsoring về một ngưỡng an toàn hơn

```python
upper_limit = df['absences'].quantile(0.95) # Tìm ngưỡng 95%
df['absences'] = df['absences'].clip(upper=upper_limit)   # Đặt ngưỡng trần cho các giá trị là upper limit

print(f"Upper limit: {upper_limit}")
```

```python
# Lưu file đã xử lý xong
df.to_csv("../data/processed/student-clean.csv")
```
