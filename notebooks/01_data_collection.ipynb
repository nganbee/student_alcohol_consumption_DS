## **Combining dataset**
Do yêu cầu dataset phải có tối thiểu 1000 dòng $\rightarrow$ nhóm quyết định gộp 2 file `.csv` trong tập dataset **Student Alcohol Consumption** gồm `student-mat.csv` và `student-por.csv` (2 file đều có chung các thuộc tính) thành `student-combine.csv` (có thêm thuộc tính `subject` cho `mat` và `por`) để xử lý cho các bước kế tiếp

```python
import pandas as pd

df_mat = pd.read_csv("../data/raw/student-mat.csv")
df_por = pd.read_csv("../data/raw/student-por.csv")
```

```python
# Thêm giá trị cho thuộc tính subject
df_mat['subject'] = 'mat'
df_por['subject'] = 'por'
```

```python
# Gộp 2 data frame
df_combine = pd.concat([df_mat, df_por], axis=0, ignore_index=True)

df_combine.to_csv("../data/raw/student-combine.csv", index=False)
```

## **How was this data collected**
Dữ liệu được thu thập từ một nghiên cứu về hành vi tiêu thụ rượu của học sinh trung học ở Bồ Đào Nha. Dữ liệu bao gồm thông tin về đặc điểm cá nhân, học tập và hành vi tiêu thụ rượu của học sinh. Dữ liệu được thu thập thông qua bảng câu hỏi tự báo cáo, nơi học sinh cung cấp thông tin về bản thân và thói quen uống rượu của họ. Nghiên cứu này nhằm mục đích hiểu rõ hơn về các yếu tố ảnh hưởng đến việc tiêu thụ rượu trong giới trẻ và các tác động tiềm ẩn của nó đối với thành tích học tập và sức khỏe.
